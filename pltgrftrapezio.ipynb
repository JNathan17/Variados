```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2.971 * np.exp(0.138 * x)
```

```python
def calcular_area_trapezio(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    area = 0
    for i in range(n):
        area += (y[i] + y[i+1]) * h / 2
    return area
```

```python
# Parâmetros
limite_inferior = 40
limite_superior = 60
```

```python
# Plotagem gráfica
x = np.linspace(1, 100, 1000)  # Intervalo de 1 a 100 dias
y = f(x)
```

```python
plt.plot(x, y, label='Curva exponencial')
plt.show()
```

```python
x_fill = np.linspace(limite_inferior, limite_superior, 100)
y_fill = f(x_fill)
for i in range(len(x_fill)-1):
    plt.fill([x_fill[i], x_fill[i], x_fill[i+1], x_fill[i+1]], [0, y_fill[i], y_fill[i+1], 0], color='orange', alpha=0.5)
    
    # Cálculo da área
area = calcular_area_trapezio(limite_inferior, limite_superior, n=1000)
print("Área aproximada sob a curva:", area)
```
